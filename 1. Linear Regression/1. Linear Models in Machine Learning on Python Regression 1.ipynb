## Linear Regression

Linear regression is probably the most fundamental machine learning algorithm in which we are trying to predict the value of a continuous variable y by using the linear combination of our predictor values X and theta values which are also called weights. We will use an optimization algorithm to find the best weights that minimizes our prediction error. It's generally much more hard to imagine how this steps work, basically imagine there are some point on a graph like the one in the picture A. Afterwards, we fit a line to the data (B) and then we optimize this line so that it has the least error while predicting our points (C). ## Under the Hood Mathematics for Linear Regression

 Firstly, we need an hypothesis which is the function that we will use for prediction. Here we have X and $\theta_1$ values which I previously mentioned. Moreover, we also have $ \theta _0 $ which is called bias or interception term.
 $$Hypothesis: h_\theta \left(x\right)=\theta _0+\theta _1x $$
 
When hypothesis tries to predict a continuos function we need another function that shows us how good or bad our hypothesis is predicting. This function is called objective function, in the context of linear regression or most Machine Learning algorithms it is also called Cost/Lost Function (In fact Cost/Lost function is another name for objective function when we want to minimize the error). For linear regression we generally use MSE(Mean Square Error) or RMSE(Root Mean Square Error), here I only present MSE.
 
 $$ Cost\  Function: J\left(\theta _0,\theta _1\right)=\frac{1}{2m}\sum _{i=1}^m\left(h_\theta \left(x^{\left(i\right)}\right)-y^{\left(i\right)}\right)^2$$
 
 But what will we do when our algorithms performs bad ? In other words, how will we make it to predict better ? When this question arises we need an optimization algorithm which is used to update our $\theta$ values so that we get a lower MSE. To do that we firstly need derivative of the cost function.
 $$\\\frac{\partial}{\partial \theta _j}\:\left(\frac{1}{2m}\sum \:_{i=1}^m\left(h_{\theta}\left(x^{\left(i\right)}\right)-y^{\left(i\right)}\right)^2\right) =  \frac{1}{m}\:\cdot \:\sum \:_{i=1}^m\left(h_{\theta }\left(x^{\left(i\right)}\right)-y^{\left(i\right)}\right)^2\cdot \:x_j^{\left(i\right)}$$
 
 We will use the optimization algorithm Gradient Descent. Gradient descent is an iterative optimization algorithm which tries to find local or global minima in every iteration by updating our $\theta$ values. Here we also have a parameter called learning rate which is represented by $\alpha$. This parameter shows step size of gradient descent in each iteration, in other words, it shows how much our $\theta$ parameters will change in each step.
 
 For updating $\theta_0$ until convergence
 $$\theta _0:=\theta\:_0-\:\alpha \:\left(\frac{1}{m}\sum _{i=1}^m\:\left(h_{\theta }\left(x^{\left(i\right)}\right)-y^{\left(i\right)}\right)\right)$$
 For updating $\theta_1$ until convergence
 $$\theta _1:=\theta _1-\:\alpha \:\left(\frac{1}{m}\sum _{i=1}^m\:\left(h_{\theta }\left(x^{\left(i\right)}\right)-y^{\left(i\right)}\right)^2\right)\cdot \:x^{\left(i\right)}$$
 
What if we have a multivariate model ? In other words, what will happen if we have more predictors than one ? In this case the algorithm changes slightly.

$$x=\begin{pmatrix}x_0\\ x_1\\ .\\ .\\ .\\ x_n\end{pmatrix} \in \:\:R^{n+1} \ \ \  \ \ \theta =\begin{pmatrix}\theta _0\\ \theta _1\\ .\\ .\\ .\\ \theta _n\end{pmatrix} \in \:\:R^{n+1}$$

$$Cost \ Function = \ \frac{1}{2m}\cdot \sum _{i=1}^m\left(\left(\sum_{j=1}^{\:n}\:\theta \:_j\cdot \:x_j^{\left(i\right)}\right)-y^{\left(i\right)}\right)^2$$

$$Gradient \ Descent \ \theta _j:=\theta _j-\:\alpha \:\cdot \frac{\partial }{\partial \theta _j}J\left(\theta \:\right) {\bf =} \:\: \theta _j:=\theta _j-\:\alpha \:\left(\frac{1}{m}\:\sum _{i=1}^m\left(h_{\theta }\:\left(x^{\left(i\right)}\right)-y^{\left(i\right)}\right)^2\cdot \:x_j^{\left(i\right)}\right)\:$$

The gradient descent equation for multivariate model and univariate model are actually the same but when n changes some terms cancel out. For instance, we have $x^{\left(i\right)}_{0\:} = 1$ for $\theta _0$. Therefore, when we have independent variable more than one we should also take into account sub script j. Furthermore, there are also different versions of Gradient Descent Batch or Mini-batch or Stochastic which i didn't mention here but by clicking the link below you can find a very good explanation of these variants of Gradient descent. Also note that, we can also have vectorized versions of these formulas which i didn't mention here and we can also use normal equation instead of gradient descent. One important drawback of normal equation is that it gets very slow when n (training instances) is large. Last but not least, our cost function is infact a special case of Maximum likelihood estimation under the normal error assumption, you can find more information about this on websites below but I will also show that in my Generalized Linear Models Notebook. ### Normality

For most of the time if our target variable is normally distributed we get better results. If It's not normally distributed we can still make it normally distributed by using some transformations such as log, square and square root.

```python
housing.SalePrice.hist(); #let's look at our target variable without any transformation used.
```

Now I will use normaltest() function to see how normally distributed our data is. This test will give us a p-value. If the p-value is bigger than 0.5 we say that distribution is normal. Documentation for `normaltest()` [link](https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.normaltest.html)

```python
normaltest(housing.SalePrice.values)
```

Our p-value is extremely low. Let's see whether log transformation makes it better.

```python
plt.hist(np.log(housing.SalePrice)); 
```

We can also directly apply log transformation to SalePrice.

The log transform can transform data that is significantly skewed right to be more normally distributed:

```python
log_saleprice = np.log(housing.SalePrice)
```

```python
log_saleprice.hist();
```

```python
normaltest(log_saleprice)
```

The p value is better but it's still not sufficient. Let's apply square root transformation.

```python
square_saleprice=np.sqrt(housing.SalePrice)
```

```python
square_saleprice.hist()
```

```python
normaltest(square_saleprice)
```

For both of the transformations we couldn't reach the normal distribution. In this sort of situations we can use another transformation called box cox. This transformation tries to get a distribution as close to a normal distribution as possible. 


$$ \text{boxcox}(y_i) = \frac{y_i^{\lambda} - 1}{\lambda} $$

```python
boxcox_transform = boxcox(housing.SalePrice)
boxcox_SalePrice = boxcox_transform[0]
lam = boxcox_transform[1]
```

```python
lam
```

```python
plt.hist(boxcox_SalePrice);
```

```python
normaltest(boxcox_SalePrice)
```

### Regression Example 1 - Summing up the things learnt up to here

Here I will construct a basic linear regression model. To begin with, let's declare predictors and target variable.

```python
predictors = housing.loc[:,['Lot Area', 'Overall Qual', 'Overall Cond', 'Year Built', 'Year Remod/Add', 'Gr Liv Area', 
'Full Bath', 'Bedroom AbvGr', 'Fireplaces', 'Garage Cars','Garage Area','Garage Qual_Fa','Sale Type_CWD','Fence_GdWo',
'Garage Qual_TA','Garage Cond_Fa','Sale Type_New','Sale Condition_Family','Land Contour_HLS','Bldg Type_Duplex',
                            'Bldg Type_TwnhsE','House Style_2.5Unf','Sale Condition_Normal']]
target = housing['SalePrice']
```

```python
lr = LinearRegression()
```

Instead of linear regression, I will use polynomial regression which means that I will have polynomial terms in hypothesis. Furthermore, this code also adds some interactions terms. One of the most important steps in feature engineering is making new terms by using combinations of other columns so that we have predictors which are highly correlated with target variable. This combinations are called interaction terms. I didn't want to spend much time on this because this notebook already will be a long one. However, I want to share some helpful resources with you [Polynomial Regression](https://towardsdatascience.com/polynomial-regression-bbe8b9d97491) [Interaction Terms](https://medium.com/@haahussain/interaction-terms-in-regression-models-4a8f1e503d49)

$$Hypothesis: h_{\theta \:}\left(x\right)=\theta _0+\theta _1x+\theta _2x^2$$


By setting `interactions_only=True` we can only create interaction terms and exclude polynomial terms. Moreover, `include_bias=False` means we exclude $\theta_0$. To learn more about parameters of this function visit the [link](https://scikit-learn.org/stable/modules/generated/sklearn.linear_model.LinearRegression.html).

```python
poly_features = PolynomialFeatures(degree=2, include_bias=True)
X_poly = poly_features.fit_transform(predictors)#adding terms to our data
```

I will use `train_test_split()` function to partition our data. This is a concept called validation which simply means that we want to train our model by using some part of our data and then use this model on new data so that we can decide whether or not the model generalizes well. You can find more about this topic on the [website](https://machinelearningmastery.com/train-test-split-for-evaluating-machine-learning-algorithms/).



```python
X_train, X_test, y_train, y_test = train_test_split(X_poly, target, test_size=0.3, random_state=42)
```

I will use StandartScaler to scale the values. Scaling is generally important for most of the algorithms. When scale of predictors are different we may end up having unstable, large weights which leads to gradient descent to converge slower. That's why, when we scale we can speed up gradient descent, moreover, it can also converge to a better point which means we have a lower error rate.
$$Standardization = \frac{X-\mu }{\sigma }$$
We subract mean from every values of X then divide by standart deviation. By doing so, the data will have zero mean and unit variance. Unlike mix-max scaling, standardization doesn't bound values between 0 and 1. Furthermore, we will use the mean and standart deviation learnt from training data to transform our test data as well.

```python
s = StandardScaler()
X_s = s.fit_transform(X_train)
```

Now let's get out target variable  as close to a normal distribution as possible by using `boxcox()` function we have used previously.

```python
boxcox2 = boxcox(y_train)
y_boxcox = boxcox2[0]
lam2 = boxcox2[1]
```

Time to fit linear regression.

```python
lr.fit(X_s, y_boxcox) # fitting linear regression.
X_test_scaled = s.transform(X_test) # We also need to scale our test values. 
# We only use transform() method because we use the mean and standart deviation calculated from train data.
y_pred = lr.predict(X_test_scaled) # predicting target
```

Now we need to do an inverse tranform to our target variable and then I will use some performance metrics to see the results of our fit. There are several metrics for regression Mean Absolute Error(MAE), Mean Square Error(MSE), Root Mean Square Error(RMSE) and $R^2 Score $. For the first example I want to be as simple as possible with these metrics, that's why, I used $R^2 \ Score $. $R^2$ shows the variability in data explain by our regression model. It is between 0 and 1, also calculated by the formula given below. In this formula SSR is the sum of squared regression and SST is the sum of squared total.$$R^2=\frac{SSR}{SST}$$

```python
y_pred_tran = inv_boxcox(y_pred,lam2)
r2_score(y_pred_tran,y_test)
```

What if we didn't get our target variable as close to a normal distribution as possible. Would that effect our performance metric ?

```python
lr = LinearRegression()
lr.fit(X_s,y_train)
lr_pred = lr.predict(X_test_scaled)
r2_score(lr_pred,y_test)
```

It definetely effects to transform our target variable. The change in the metric is slight but important. Let's see how our model predicts the data visually. I will use seaborn for that purpose, afterwards, I will also plot a classical regression line.

```python
plt.figure(figsize=(10,8))
sns.set_context('talk')
sns.set_style('ticks')
sns.set_palette('dark')
ax = plt.axes()

# I'm going to use y_test and y_pred here.
ax.scatter(y_test, y_pred, alpha=.5,color='violet')
```

We can also plot the classical regression line as shown below.

```python
plt.figure(figsize=(10,8))
sns.set()
plt.scatter(y_pred_tran,y_test,marker = '+',c='black')
plt.xlabel('Predicted values')
plt.ylabel('Actual values')
#Plots x=y line
plt.plot([0,np.max(y_test)],[0,np.max(y_test)], c = 'r');
```

Even though the difference isn't much there is still a difference when we transform our target variable into normal distribution.
